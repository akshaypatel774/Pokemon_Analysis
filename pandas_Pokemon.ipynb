# __Make changes to data__
__[For More Markdown](https://www.ibm.com/docs/en/watson-studio-local/1.2.3?topic=notebooks-markdown-jupyter-cheatsheet)__

```python
#[:, 4:10] 1st colon for rows and 2nd for columns
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
df.head(3)
```

```python
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]]+cols[4:12]]
df.head(5)
```

```python
#Save file with no index
df.to_csv('Updated_Pokemon.csv', index = False)
df.to_csv('Updated_Pokemon.txt', index=False, sep='\t')
```

# Filtering

```python
Grass = df.loc[df['Type 1'] == 'Grass']
Grass.head(5)
```

```python
# Can also use & as|
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
new_df
```

```python
# add new index and drop older one
new_df.reset_index(drop=True, inplace=True)
new_df
```

```python
# HP > 100 and (Poison or grass type (in 2 Columns))
new_df1 = df.loc[((df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')) & (df['HP'] > 100)]
new_df1
```

```python
# HP > 100 and (Fire or grass type (in same 1 column))
df.loc[df['Type 1'].str.contains('Grass|Fire') & (df['HP'] > 100)]
```

```python
# Names starting with Pi (^ means starting)
df.loc[df['Name'].str.contains('^Pi')]
```

```python
# Change in Type 1 flamer as fire
df.loc[df['Type 1'] == 'Flamer', 'Type 1'] = 'Fire'
df
```
